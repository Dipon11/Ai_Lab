```python
import random

movies = [
    {"title": "The Godfather", "hint": "It's a gangster movie made in Hollywood. It's an action-packed film."},
    {"title": "The Hangover", "hint": "Three buddies wake up from a bachelor party in Las Vegas, with no memory of the previous night and the bachelor missing. They make their way around the city in order to find their friend before his wedding."},
    {"title": "Inception", "hint": "A thief who steals corporate secrets through the use of dream-sharing technology is given the inverse task of planting an idea into the mind of a C.E.O., but his tragic past may doom the project and his team to disaster."},
    {"title": "Titanic", "hint": "A ship crashes, Leo dies for no reason."},
    {"title": "Jurassic Park", "hint": "Movie about dinosaurs."},
]

def hangman():
    chosen_movie = random.choice(movies)
    movie_title = chosen_movie["title"].lower()
    hint = chosen_movie["hint"]
    guessed_letters = []
    attempts = 6
    mistake_count = 0



    while True:
        print("\nAttempts remaining:", attempts)
        print_hangman(attempts)
        print("\nGuessed letters:", guessed_letters)

        display_movie = ""
        for letter in movie_title:
            if letter in guessed_letters or letter == " ":
                display_movie += letter
            else:
                display_movie += "_ "
        print(display_movie)

        guess = input("Enter a letter or guess the full movie title: ").lower()

        if guess in guessed_letters:
            print("You've already guessed that letter. Try again.")
            continue

        if len(guess) > 1:
            if guess == movie_title:
                print("\nCongratulations! You won! The movie was:", movie_title.title())
                break
            else:
                print("Wrong guess!")
                attempts -= 1
                mistake_count += 1
                if mistake_count == 3 or mistake_count == 4:
                    print("Hint:", hint)
                continue

        guessed_letters.append(guess)

        if guess in movie_title:
            print("Correct guess!")
        else:
            print("Wrong guess!")
            attempts -= 1
            mistake_count += 1
            if mistake_count == 3 or mistake_count == 4:
                print("Hint:", hint)

        if all(letter in guessed_letters or letter == " " for letter in movie_title):
            print("\nCongratulations! You won! The movie was:", movie_title.title())
            break
        elif attempts == 0:
            print("\nGame over! You lost. The movie was:", movie_title.title())
            break

        # AI player's turn
        ai_guess = ai_player(movie_title, guessed_letters)
        print("AI guessed:", ai_guess)

        if ai_guess in movie_title:
            print("AI guessed a correct letter!")
        else:
            print("AI guessed wrong.")
            attempts -= 1

def ai_player(movie_title, guessed_letters):
    candidates = [letter for letter in movie_title if letter not in guessed_letters]
    ai_guess = ""

    if candidates:
        ai_guess = random.choice(candidates)

    if ai_guess == "":
        ai_guess = random.choice(list("abcdefghijklmnopqrstuvwxyz"))

    return ai_guess

def print_hangman(attempts):
    stages = [
        """
            --------
            |      |
            |      O
            |     /|\\
            |      |
            |     / \\
            -
        """,
        """
            --------
            |      |
            |      O
            |     /|\\
            |      |
            |     /
            -
        """,
        """
            --------
            |      |
            |      O
            |     /|\\
            |      |
            |
            -
        """,
        """
            --------
            |      |
            |      O
            |     /|
            |      |
            |
            -
        """,
        """
            --------
            |      |
            |      O
            |      |
            |      |
            |
            -
        """,
        """
            --------
            |      |
            |      O
            |
            |
            |
            -
        """,
        """
            --------
            |      |
            |
            |
            |
            |
            -
        """
    ]
    print(stages[attempts])

# Run the Hangman game
hangman()
```
